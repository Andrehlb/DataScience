```python
#os índices no Python vão de 0 até n-1
#onde n é o tamanho da dimensão: 1D, 2D, 3D
x = np.linspace(start=1, stop=10, num=2)
print('x:', x)
print('shape:', x.shape)
```

```python
#os índices no Python vão de 0 até n-1
#onde n é o tamanho da dimensão: 1D, 2D, 3D...
x = np.linspace(start=10, stop=100, num=10)
print('x:', x)
print('shape:', x.shape)
```

```python
#Extração de elementos 
#acessando arrays unidimensionais
print('x: ', x)
print('Primeiro elemento', x[0])
print('Segundo elemento', x[1])
print('Último elemento', x[9])
print('Último elemento', x[-1])
```
